### #2 Preprocessing Data

#### Step-1: Data Cleaning

```python
# periksa apakah ada null data atau tidak
null_data = dataset.isna().sum()
print(null_data)

# berapa total data null yang dimiliki
print('Data kosong yang dimiliki :',null_data.sum())
```

```python
# karena data yang kosong cukup sedikit, kita dapat menghapus data tersebut.
# pada hal ini saya akan coba 2 kondisi, kondisi pertama data kosong saya hapus
# kemudian kondisi kedua, data kosong saya inputkan dataset.median()

data_drop = dataset.dropna()
data_median = dataset.fillna(dataset.median())
```

```python
# cek kembali dimensi data

print('Data awal :', dataset.shape)
print('Data drop :', data_drop.shape)
print('Data median :', data_median.shape)
```

#### Part-2: Transformasi Data

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# perlu dilakukan transform scaling karena data yang kita miliki mempunya beragam batasan atas dan bawah
# maka perlu di scaling sehingga interval nilai menjadi (0, 1)
# kemudian, perlu dilakukan transform LabelEncoder karena machine learning tidak dapat mengeksekusi tipe data object/str 
# maka dilakukan transformasi tersebut.
```

```python
scaling = MinMaxScaler()
LE = LabelEncoder()
```

```python
# Data_DROP
data_int = data_drop.drop(['Month','VisitorType','Weekend','Revenue'], axis=1)
column = []
for i in data_int:
    column += [i]

data_drop[column] = scaling.fit_transform(data_drop[column])
```

```python
# DATA DROP

data_drop['Month'] = LE.fit_transform(data_drop['Month'])
data_drop['VisitorType'] = LE.fit_transform(data_drop['VisitorType'])
```

```python
# DATA MEDIAN
data_int_median = data_median.drop(['Month','VisitorType','Weekend','Revenue'],axis=1)

column_int = []
for i in data_int_median:
    column_int += [i]
    
data_median[column_int] = scaling.fit_transform(data_median[column_int])
```

```python
# Data MEDIAN

data_median['Month'] = LE.fit_transform(data_median['Month'])
data_median['VisitorType'] = LE.fit_transform(data_median['VisitorType'])
```

#### Part-3: Pisah data target (y) dan data feature (x)

```python
# DATA DROP
x_drop = data_drop.drop(['Revenue'], axis=1)
y_drop = data_drop['Revenue']

# DATA MEDIAN
x_med = data_median.drop(['Revenue'], axis=1)
y_med = data_median['Revenue']
```

### #3 Model Machine Learning

#### Part-1: pisah menjadi 80:20 

```python
from sklearn.model_selection import train_test_split
```

```python
# DATA DROP

x_drop_train, x_drop_test, y_drop_train, y_drop_test = train_test_split(x_drop, y_drop, test_size=0.2, random_state=0)
```

```python
# DATA MEDIAN

x_med_train, x_med_test, y_med_train, y_med_test = train_test_split(x_med, y_med, test_size=0.2, random_state=0)
```

### #4 Metode Machine Learning

```python
# terdapat berbagai macam metode, dan metode yang digunakan disesuaikan dengan kasus dari dataset kita
# untuk melihat keakuratan dari hasil prediksi, dapat dengan menggunakan methode .score( ) 
# untuk lebih luas dapat dilihat dengan mengimport libary berikut
from sklearn.metrics import confusion_matrix, classification_report
```

#### Metode Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
```

```python
# DATA DROP
logreg_drop = logreg.fit(x_drop_train, y_drop_train)

y_drop_pred = logreg_drop.predict(x_drop_test)

# DATA MEDIAN
logreg_med = logreg.fit(x_med_train, y_med_train)

y_med_pred = logreg_med.predict(x_med_test)
```

```python
# akurasi DATA DROP
print('Training Accuracy :', logreg_drop.score(x_drop_train, y_drop_train))
print('Testing Accuracy :', logreg_drop.score(x_drop_test, y_drop_test))

print('\nConfusion Matrix :\n', confusion_matrix(y_drop_test, y_drop_pred, labels=[1,0]))
print('\nClassification Report :\n', classification_report(y_drop_test, y_drop_pred))
```

```python
# akurasi DATA MEDIAN
print('Training Accuracy :', logreg_med.score(x_med_train, y_med_train))
print('Testing Accuracy :', logreg_med.score(x_med_test, y_med_test))

print('\nConfusion Matrix :\n', confusion_matrix(y_med_test, y_med_pred, labels=[1,0]))
print('\nClassification Report :\n', classification_report(y_med_test, y_med_pred))
```

### #5 Kesimpulan

1. Dalam kondisi ini, Training Accuracy menggunakan data_drop lebih rendah dibandingkan Training Accuracy menggunakan data_median hal ini dapat terjadi karena pada data_drop kita    memiliki data yang lebih sedikit dibanding pada data_median. 
2. Apabila kita  bandingkan kesalahan prediksi ketika menggunakan data_drop dan data_median adalah sebagai berikut.
   FP : data_drop < data_median
   FN : data_drop < data_median
3. Apabila kita bandingan prediksi benar ketika menggunakan data_drop dan data_median adalah sebagai berikut.
   TP : data_drop > data_median
   TN : data_drop > data_median
4. Maka perbandingan Classification_Report ketika menggunakan data_drop dan data_median adalah sebagai berikut.
   -  Precision
      - False : data_drop < data_median
      - True  : data_drop > data_median
   -  Recall
      - False : data_drop > data_median
      - True  : data_drop > data_median
   -  f1-Score
      - False : data_drop > data_median      
      - True  : data_drop > data_median
   -  Support
      - False : data_drop > data_median
      - True  : data_drop > data_median
